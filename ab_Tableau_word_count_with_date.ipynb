```python
import pandas as pd
from collections import Counter
```

```python
filepath = "Resources/dirty_Tableau_pre_covid_sentiments_predictions_headlines_lr_model.csv"
df = pd.read_csv(filepath)
```

```python
# df.columns
df.drop(["Unnamed: 0"], axis = 1, inplace = True)
df
```

```python
filepath2 = "Resources/precovid_articles_dates.csv"
df1 = pd.read_csv(filepath2, header=None)
df2 = df1.rename(columns={0:"text", 1:"date"})
df2
```

```python
merged_df = pd.merge(df, df2, on="text", how="inner")
merged_df
```

```python
def convert(lst):
    return lst.split()
```

```python
merged_df["text_list"] = merged_df.apply(lambda row: convert(row["text"]), axis=1)
merged_df
```

```python
df_textlist = merged_df[["predicted_label", "date", "text_list"]]
df_textlist
```

```python
out = []
for index, row in df_textlist.iterrows():
    for item in row["text_list"]:
        row["keywords"] = item
        out += [row.copy()]
        
flattened_df = pd.DataFrame(out)
flattened_df
```

```python
flattened_df1 = flattened_df.drop("text_list", axis=1)
flattened_df1
```

```python
# Removing Numeric Values from Keywords
flattened_df2 = flattened_df1[~flattened_df1["keywords"].str.contains("^\d+$")]
flattened_df2
```

```python
# Cleaning text with symbols
import re
flattened_df2["keywords_clean"] = flattened_df2["keywords"].map(lambda x: re.sub(r'\W+', '', x))
```

```python
flattened_df3 = flattened_df2[["predicted_label", "date", "keywords_clean"]]
flattened_df3
```

```python
# Check for empty rows
import numpy as np
np.where(pd.isnull(flattened_df3))
```

```python
# Filter out negative sentiment
neg_df = flattened_df3.loc[flattened_df3["predicted_label"] == "Negative"]
neg_df
```

```python
x = neg_df["keywords_clean"].value_counts().to_frame()
x
```
