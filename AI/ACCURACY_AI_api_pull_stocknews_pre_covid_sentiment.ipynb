```python
date_s = pd.date_range(start=start_date,end=end_date)
date_range = date_s.strftime("%m%d%Y")
date_range
```

```python
ranges = [i + "-" + i for i in date_range]
```

```python
api_url = "https://stocknewsapi.com/api/v1/category?section=alltickers&items=50&token=0mrpl4lmd8xwbtix2fbgtbtqtd8hlhaeraxxx4xr&date="
```

```python
full_api_dates = [api_url + i for i in ranges]
full_api_dates
```

```python
def api_call(x):
    articles = requests.get(x).json()
    articles_list = articles["data"]
    
    articles_x = []
    for articles in articles_list:
        articles_x.append(articles["title"])
    return articles_x
```

```python
api_scrape = [api_call(i) for i in full_api_dates]
api_scrape
```

```python
# len(api_scrape)
```

```python
def flatten_list(nested_list):
    x = []
    for i in nested_list:
        x.extend(i)
    return x
```

```python
articles_dates = flatten_list(api_scrape)
articles_dates
```

```python
# len(articles_dates)
```

```python
def api_call_sentiment(y):
    articles = requests.get(y).json()
    articles_list = articles["data"]
    
    articles_x = []
#     pprint(articles)

    for articles in articles_list:
        articles_x.append(articles["sentiment"])
    return articles_x
```

```python
api_scrape1 = [api_call_sentiment(j) for j in full_api_dates]
```

```python
test_news_sentiment = flatten_list(api_scrape1)
```

```python
df_sentiment = pd.DataFrame(test_news_sentiment, articles_dates)
df_sentiment1 = df_sentiment.reset_index()
df_sentiment1
```
