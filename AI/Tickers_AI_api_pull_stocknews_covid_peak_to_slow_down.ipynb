```python
def api_call(x):
    articles = requests.get(x).json()
    articles_list = articles["data"]
    
    articles_x = []
    for articles in articles_list:
        articles_x.append(articles["title"])
    return articles_x
```

```python
api_scrape = [api_call(i) for i in full_api_dates]
api_scrape
```

```python
# len(api_scrape)
```

```python
def flatten_list(nested_list):
    x = []
    for i in nested_list:
        x.extend(i)
    return x
```

```python
articles_dates = flatten_list(api_scrape)
articles_dates
```
